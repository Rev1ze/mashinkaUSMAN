1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

2. Import the dataset US_Baby_Names_right.csv

```python
baby_names = pd.read_csv('US_Baby_Names_right.csv')
```

3. Assign it to a variable called baby_names

```python
baby_names.info()
```

4. See the first 10 entries

```python
baby_names.head(10)
```

5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names.drop(axis=1, columns=['Unnamed: 0', 'Id'], inplace=True)
```

```python
baby_names
```

6. Are there more male or female names in the dataset?

```python
baby_names[['Gender', 'Count']].groupby('Gender').count()
```

7. Group the dataset by name and assign to names

```python
baby_names[['Name', 'Count']].groupby('Name').sum().head(5)
```

```python
baby_names[['Name', 'Count']].groupby('Name').sum().sort_values('Count', ascending=False).head(5)
```

8. How many different names exist in the dataset?

```python
len(baby_names.Name.unique())
```

```python
baby_names.Name.nunique()
```

9. What is the name with most occurrences?

```python
baby_names[['Name', 'Count']].groupby('Name').sum().sort_values('Count', ascending=False).head(1).reset_index().Name[0]
```

10. How many different names have the least occurrences?

```python
least_name = baby_names[['Name', 'Count']].groupby('Name').sum()
```

```python
least_name[least_name.Count == least_name.Count.min()]
```

```python
len(least_name[least_name.Count == least_name.Count.min()])
```

11. What is the median name occurrence?

```python
median_name = baby_names[['Name', 'Count']].groupby('Name').sum()
```

```python
median_name[median_name['Count'] == median_name['Count'].median()]
```
